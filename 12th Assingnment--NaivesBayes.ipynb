```python
from sklearn.model_selection import train_test_split

salary_train,salary_test = train_test_split(data,test_size=0.3)
```

```python
salary_train
```

```python
 from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer()
trn_salary = count_vec.fit_transform(salary_train.Salary)
tst_salary = count_vec.transform(salary_test.Salary)
```

```python
trn_salary.shape, tst_salary.shape , salary_train.shape
```

```python
trn_salary.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
import numpy as np

classifier_mb = MB()
classifier_mb.fit(trn_salary, salary_train.Salary)

# Model Test
train_pred_m = classifier_mb.predict(tst_salary)

# Model Accuracy over test
accuracy_test = np.mean(train_pred_m==salary_test.Salary)
accuracy_test
```

```python
# Model Accuracy over train
test_pred_m = classifier_mb.predict(trn_salary)
accuracy_train = np.mean(test_pred_m==salary_train.Salary)
accuracy_train
```
